```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.stats import gamma 
```

```python
x = np.linspace(0,20,1000)
Q2 = [1, 2, 5, 10]
fig, axes = plt.subplots(1,1, figsize=(6,4))

for q2 in Q2:
    y = gamma.pdf(x, 1, loc=0, scale=q2)
    axes.plot(x, y/y.max(), label=fr'$q_2$ = {q2}')
    
axes.set_ylabel('Probability density (normalised)')
axes.set_xlabel(r'$\phi$')
axes.set_title(r'Gamma distributions ($q_1$ = 1)')
axes.legend()
plt.savefig('gamma_q2_plot.png', dpi=250, bbox_inches='tight')
plt.show()
```
